```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the dataset
df = pd.read_csv(r"C:\Users\theer\Downloads\house_price.csv")

# Data Exploration
print("Unique values in each feature:")
for column in df.columns:
    print(column, ":", df[column].unique())
    print("Length of unique values:", len(df[column].unique()))

# Perform statistical analysis
print("\nStatistical analysis:")
print(df.describe())

# Data Cleaning
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Data Analysis
# Filter the data based on conditions
filtered_data = df[(df["bath"] > 2) & (df["price"] < 5000)]

# Plot chart with relevant features
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data["bath"], filtered_data["price"])
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.title("Bathrooms vs Price")
plt.show()

# Count number of properties from each location and represent it visually
plt.figure(figsize=(10, 6))
sns.countplot(x="location", data=df)
plt.title("Number of Properties in Each Location")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

```
